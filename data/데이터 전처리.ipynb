```python
# 라이브러리 import
import numpy as np
import pandas as pd
```

```python
# 파일 불러오기
name = pd.read_csv("pokemon_data/pokemon_name.csv")
data = pd.read_csv("pokemon_data/pokemon.csv")
data.head()
```

```python
data.columns
```

```python
# 불필요한 column drop
data.drop(['Column1', 'german_name', 'japanese_name', 'species', 'abilities_number', 'ability_1', 'ability_2',
       'ability_hidden', 'catch_rate', 'base_friendship',
       'base_experience', 'growth_rate', 'egg_type_number', 'egg_type_1',
       'egg_type_2', 'percentage_male', 'egg_cycles'], axis=1, inplace=True)
data.tail()
```

```python
# 1세대 포켓몬만 추리기
data = data[data["generation"] == 1]
data.tail()
```

```python
# 이젠 모든 포켓몬이 1세대이므로 column drop
data.drop(['generation'], axis=1, inplace=True)
data.head()
```

```python
# 메가진화 버전 포켓몬 거르기
data = data[data["name"].str.startswith("Mega") == False]
data.tail()
```

```python
# 영어이름 기준으로 한글 이름 추가
data = pd.merge(name, data, how='inner', left_on='Name', right_on='name')
data
```

```python
# 중복되는 column drop
data.drop(['번호', 'Name', 'name'], axis=1, inplace=True)
data.head()
```

```python
# 추가된 이름 column 이름 변경
data.rename(columns = {'이름':'name'},inplace=True)
data.tail()
```
